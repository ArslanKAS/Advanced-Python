A word cloud is a visual representation of text data, which is often used to depict keyword metadata on websites, or to visualize free form text.

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
text = """
"Python"
"Convolutional Neural Networks (CNN)"
"Recurrent Neural Networks (RNN)"
"Long Short-Term Memory (LSTM)"
"Generative Adversarial Networks (GAN)"
"Transformer Networks"
"Deep Q-Networks (DQN)"
"Autoencoder"
"Restricted Boltzmann Machine (RBM)"
"Random Forest"
"Gradient Boosting Machine (GBM)"
"XGBoost"
"LightGBM"
"Facebook's DeepFace"
"LinkedIn's LSTM-based recommendation algorithm"
"Instagram's DeepLink"
"Twitter's text-based DeepMoji"
"Facebook's DeepText"
"LinkedIn's Deep Learning based search ranking system"
"Instagram's DeepFocus"
"Twitter's DeepRec".
"""
```

## **Facebook Shape**

```python
from wordcloud import WordCloud

mask = np.array(Image.open("facebook.png"))

fb = WordCloud(background_color="white", mask=mask,
                contour_width=7).generate(text)
                
plt.imshow(fb, interpolation='bilinear')
plt.axis("off")
```
